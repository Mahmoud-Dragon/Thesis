```python
for x in ['grade of the house']:
    q75,q25 = np.percentile(df.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    df.loc[df[x] < min,x] = np.nan
    df.loc[df[x] > max,x] = np.nan
    # The instances above q75+(1.5*intr_qr) or below q25-(1.5*intr_qr) are located and replaced by NaN
```

```python
numerical_columns = ['Postal Code','living area','living_area_renov']


df[numerical_columns] = ( 
    
    df[numerical_columns] - df[numerical_columns].min()  
    
) /( 
    
    df[numerical_columns].max() - df[numerical_columns].min()
)


df.head() #Normalization of these columns as they have large values
     
```

```python
from sklearn.model_selection import train_test_split

d = df.drop('Price',axis=1) # Independent variable
e = df['Price'] # Dependent variable

d_train, d_test, e_train, e_test = train_test_split(d,e,test_size=0.2,random_state=23)

print('Training dataset size:', d_train.shape[0])
print('Testing dataset size:', d_test.shape[0])

# splitting the dataset into 0.8 training data and 0.2 testing data
```

```python
from sklearn.linear_model import LinearRegression


lm = LinearRegression()


lm.fit(d_train, e_train);
# fitting in the data in the model
```

```python
parameter = ['theta_0'] + ['theta_{}'.format(i+1) for i in range(d.shape[1])]
columns = ['y-intersect'] + list(d.columns.values)
sk_theta = pd.Series([lm.intercept_] + list(lm.coef_))

parameter_df = pd.DataFrame({'parameter':parameter,'column':columns, 'theta_value': sk_theta})
parameter_df

# Creating the linear regression equation
```

```python
y_pred = lm.predict(d_test) # predicting the prices using the testing data
```

```python
y_pred
```

```python

# Evaluation
from sklearn.metrics import mean_squared_error
rmse = mean_squared_error(y_pred, e_test, squared=False)

# R_square
R_square = lm.score(d_test,e_test)
print('The root mean square error (RMSE) is: ', rmse)
print('R^2 obtained is:', R_square)                         
```
