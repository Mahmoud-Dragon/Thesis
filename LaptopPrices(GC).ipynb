```python
df=df.drop(['Operating System Version'],axis=1)
```

```python
from sklearn.model_selection import train_test_split

d = df.drop('Price',axis=1) # Independent variable
e = df['Price'] # Dependent variable
```

```python
d_train, d_test, e_train, e_test = train_test_split(d,e,test_size=0.2,random_state=23)

print('Training dataset size:', d_train.shape[0])
print('Testing dataset size:', d_test.shape[0])
```

```python
from sklearn.tree import DecisionTreeRegressor
reg= DecisionTreeRegressor(criterion='squared_error',max_depth=4,min_samples_split=10,random_state=5)
```

```python
reg.fit(d_train,e_train)
```

```python
y_pred= reg.predict(d_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
r2_score(e_test,y_pred)
```

```python
rmse = mean_squared_error(y_pred, e_test, squared=False)
```
